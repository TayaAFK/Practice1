*  Для заказов более 5000 единиц, клиент получает скидку 15%, а использование промокода SUPERDISCOUNT увеличивает скидку до 20% (вместо 15%).

Ввод:

Оформите ваш код в виде функции

````python
cost=int(input('Введите стоимость единицы товара: '))
kolvo=int(input('Введите количество товара: '))
promo=input('Введите промокод: ')

def F(x, y, z):
  result=0
  disc=0
  if 1000<y<5000:
    if z=='SUPERDISCOUNT':
      result=(x*y)/100*90
      disc=10
    elif z!='SUPERDISCOUNT':
      result=(x*y)/100*95
      disc=5
  elif y>5000:
    if z=='SUPERDISCOUNT':
      result=(x*y)/100*80
      disc=20
    elif z!='SUPERDISCOUNT':
      result=(x*y)/100*85
      disc=15
  else:
    result=x*y
  print('Ваша скидка: ' + str(disc) + '%')
  return('Итоговая сумма: ' + str(result))

print(F(cost, kolvo, promo))
````

Этап 2:
Оформите ваш код в виде функции

Ввод:
```
apple banana apple
```

Вывод:

```
apple: 2,
banana: 1
```

````python
stroka=input()

def F(s):
  word=''
  words1=[]
  words2=[]
  NUM=[]
  for x in s:
    if x!=' ':
      word+=x
    else:
      words1.append(word)
      word=''
  words1.append(word)

  for i in range(0,len(words1)):
    c=1
    for j in range(i+1, len(words1)):
      if words1[i]==words1[j]:
        c+=1
    if words1[i] not in words2:
        words2.append(words1[i])
        NUM.append(c)

  for t in range(len(words2)):
    print(words2[t], ':', NUM[t])
  return('')
print(F(stroka))
````
