# Задание 1

**Описание:** Создайте иерархию классов для разных типов сотрудников в компании. Реализуйте родительский класс Employee и дочерние классы Manager и Developer. Каждый класс должен иметь метод для расчета зарплаты на основе различных критериев класса.


Отрабатываемый принцип: Наследование



```python
class Employee:
  def __init__(self, name, quantity):
    self.name = name
    self.quantity = quantity

  def calculate_salary(self):
    self.quantity = int(self.quantity) * 1000
    return(self.quantity)


class Manager(Employee):
  def calculate_salary1(self):
    self.quantity = self.quantity * 2
    print (f'Сотрудник {self.name} получает такую зарплату: {self.quantity}')


class Developer(Employee):
  def calculate_salary1(self):
    self.quantity = self.quantity * 5
    print (f'Сотрудник {self.name} получает такую зарплату: {self.quantity}')



name = input('Введите имя сотрудника: ')
post = input('Введите должность сотрудника: ')
quantity = input('Введите количество дней, которые сотрудник работал: ')

if post=='Менеджер' or post=='Manager':
  wages=Manager(name=name, quantity=quantity)
  wages.calculate_salary()
  print(wages.calculate_salary1())
elif post=='Разработчик' or post=='Developer':
  wages=Developer(name=name, quantity=quantity)
  wages.calculate_salary()
  print(wages.calculate_salary1())
else:
  print('Вы неправильно ввели должность')
```
