# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def unique_sym(List):
  result=[]
  for element in List:
    if element not in result:
      result.append(element)
  return result

first_numbers=input('Введите символы для первого списка черех пробел: ')
second_numbers=input('Введите символы для второго списка черех пробел: ')

first_list=first_numbers.split()
second_list=second_numbers.split()

first_list=unique_sym(first_list)
second_list=unique_sym(second_list)

result1=[]
duplicate=[]

for perebor in first_list:
  if perebor not in second_list:
    result1.append(perebor)
  else:
    duplicate.append(perebor)
for deleted in duplicate:
  second_list.remove(deleted)

result1=' '.join(result1)
result2=' '.join(second_list)

print('уникальные элементы первого списка:')
print(result1)
print('уникальные элементы второго списка::')
print(result2)
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
import requests
from PIL import Image
from IPython.display import display
import io

def first_func():
  print('Усыновим питомца!')
  name=input('Введите имя питомца: ')
  typ=input('Введите тип питомца: ')
  age=input('Введите возраст питомца: ')
  starve=5
  happy=5
  energe=5
  data=[]
  data.append(name)
  data.append(typ)
  data.append(age)
  data.append(starve)
  data.append(happy)
  data.append(energe)
  return data

def second_func(params1, params2, params3):
  while True:
    print('1. Покормить питомца')
    print('2. Поиграть с питомцем')
    print('3. Уложить питомца спать')
    print('4. Назад')
    choice=input()
    if choice=='1':
      print('Вы покормили питомца!')
      params1+=1
      params2+=1
      params3-=1
    elif choice=='2':
      print('Вы поиграли с питомцем')
      params1-=1
      params2+=2
      params3-=2
    elif choice=='3':
      print('Вы уложили питомца спать')
      params1-=2
      params2-=1
      params3+=3
    elif choice=='4':
      parametr=[params1, params2, params3]
      return parametr
      break
    else:
      print('Такой команды нет')
      break
    if params1<0 or params2<0 or params3<0:
      print('Вы плохо следили за питомцем!')
      print('Питомец умер!')
      return 0


def third_func(params1, params2, params3, params4):
  print('Состояние питомца:')
  print('Имя:', params1)
  print('Сытость:', params2)
  print('Радость:', params3)
  print('Энергия', params4)

data=first_func()
while True:
  print('1. Взаимодействовать с питомцем')
  print('2. Посмотреть состояние питомца')
  print('3. Выйти')
  choice=input()
  if choice=='1':
      result=second_func(data[3], data[4], data[5])
      if result==0:
        url='https://api.memegen.link/images/doge.png'
        result = requests.get(url)
        img = Image.open(io.BytesIO(result.content))
        display(img)
        break
      else:
        data[3]=result[0]
        data[4]=result[1]
        data[5]=result[2]
  elif choice=='2':
    third_func(data[0], data[3], data[4], data[5])
  elif choice=='3':
    break
````

Цель - создать менеджера команды Pokémon, который позволит пользователям:

Для данной задачи используйте: https://pokeapi.co/


````python
import requests
import random

def remov_pok(Dict):
  name=input('Введите имя покемона, которого хотите убрать: ')
  team_pok=[]
  for pokemon in Dict.keys():
    team_pok.append(pokemon)
  if name in team_pok:
    del Dict[name]
    print('Ваши покемоны:')
    for pok in Dict.keys():
      print(pok)
    print()
    return Dict
  else:
    print('У вас в команде нет такого покемона')
    print('Ваши покемоны:')
    for pok in Dict.keys():
      print(pok)
    print()
    return Dict

def add_pok(List, Dict):
  name=input('Введите имя покемона, которого хотите добавить: ')
  all_pok=[]
  for pokemon in List:
    all_pok.append(pokemon['name'])
  if name not in all_pok:
    print('Такого покемона нет в списке')
    print()
    return Dict
  else:
    number=all_pok.index(name)
    team_pok=[]
    for pokemon in Dict.keys():
      team_pok.append(pokemon)
    if name in team_pok:
      print('У вас в команде есть этот покемон')
      print()
      return Dict
    else:
      Dict[name]=List[number]['url']
      print('Ваши покемоны:')
      for pok in Dict.keys():
        print(pok)
      print()
      return Dict

def info_pok(Dict):
  name=input('Введите имя покемона, информацию о котором хотите узнать: ')
  if name in Dict.keys():
    URL=Dict[name]
    response=requests.get(URL)
    pokemon=response.json()['abilities']
    for perebor in pokemon:
      name_ab=perebor['ability']['name']
      print('Спобность:', name_ab)
      url_ab=perebor['ability']['url']
      response=requests.get(url_ab)
      ability=response.json()
      effects=ability['effect_entries']
      for effect in effects:
        if effect['language']['name']=='en':
          print('Эффект:', effect['effect'])
  else:
    print ("Такого покемона нет у вас в команде")

def search_pok(List):
   name=input('Введите имя покемона, информацию о котором хотите узнать: ')
   all_pok=[]
   for pokemon in List:
    all_pok.append(pokemon['name'])
   if name not in all_pok:
    print('Такого покемона нет в списке')
    print()
   else:
    number=all_pok.index(name)
    URL=List[number]['url']
    response=requests.get(URL)
    pokemon=response.json()['abilities']
    for perebor in pokemon:
      name_ab=perebor['ability']['name']
      print('Спобность:', name_ab)
      url_ab=perebor['ability']['url']
      response=requests.get(url_ab)
      ability=response.json()
      effects=ability['effect_entries']
      for effect in effects:
        if effect['language']['name']=='en':
          print('Эффект:', effect['effect'])

def battle_pok(Dict, pokemon):
  all_pok=[]
  print('Ваши покемоны:')
  for pokemons in Dict.keys():
    all_pok.append(pokemons)
    print(pokemons)
  choice=input('Выберите покемона из своей команды ')
  if choice not in all_pok:
    print('У вас в команде нет такого покемона')
  else:
    print()
    print('Против вас сражается:', pokemon)
    HP_enemy=random.randint(1,50)
    print('Его HP:', HP_enemy)
    HP_player=40
    print('Ваше HP:', HP_player)
    print('Первый ходит', pokemon)
    print()
    while True:
      print('Ход врага!')
      attack_enemy=random.randint(1, 10)
      print(f'Враг нанёс {attack_enemy} урона')
      HP_player-=attack_enemy
      print('Ваше HP:', HP_player)
      print()
      if HP_player<=0:
        break
      print('Ваш ход!')
      attack_player=random.randint(1, 12)
      print(f'Вы наносите {attack_player} урона')
      HP_enemy-=attack_player
      print('HP врага:', HP_enemy)
      print()
      if HP_enemy<=0:
        break
    if HP_enemy<=0:
      print('Вы выиграли!')
    else:
      print('Вы проиграли!')




pokemons={}
url='https://pokeapi.co/api/v2/pokemon?limit=1302'
response=requests.get(url)
LIST=response.json()["results"]
print('В вашей команде должно быть 3 покемона, напишите имена, чтобы добавить их в команду:')
while len(pokemons)!=3:
  name=input()
  count=0
  for creature in LIST:
    if creature['name']==name:
      pokemons[name]=creature['url']
      count+=1
  if count==0:
    print('Такого покемона нет в списке')
  print('Количество покемонов в команде:', len(pokemons))

number=random.randint(1, 1303)
pokemon_enemy=LIST[number]['name']

while True:
  print('Введите 1, чтобы добавить покемона в свою команду')
  print('Введите 2, чтобы удалить покемона из своей команды')
  print('Введите 3, чтобы посмотреть подробную информацию об покемоне')
  print('Введите 4, чтобы найти покемона по имени')
  print('Введите 5, чтобы устроить тренировочный бой между двумя покемонами')
  print('Введите 6, чтобы выйти')
  choice=input()
  if choice=='1':
    add_pok(LIST, pokemons)
  elif choice=='2':
    remov_pok(pokemons)
  elif choice=='3':
    info_pok(pokemons)
  elif choice=='4':
    search_pok(LIST)
  elif choice=='5':
    battle_pok(pokemons, pokemon_enemy)
  elif choice=='6':
    break
  else:
    print('Нет такой команды')
````
